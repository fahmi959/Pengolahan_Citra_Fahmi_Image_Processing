Penyajian data *Breast Cancer* 


```
import pandas as pd
import matplotlib.pyplot as plot
a="https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer/breast-cancer.data"
pre=pd.read_csv(a)
pre.columns=["class","age","menopause","tumor-size","inv-nodes","node-caps","deg-malig","breast","breast-quad","ok"]
data=pre.drop(columns=["ok"])
data.head(10)
```

#Tabel Statistik
Berikut adalah tabel statistik dari data *tumor-size*

1. Menampilkan jumlah dari : *tumor-size*

```
tabel=data.groupby(['tumor-size'])['tumor-size'].count()
print(tabel)
```

2. Menampilkan diagram batang dari : *tumor-size*

```
tabel.plot.bar(x='tumor-size', y='count', title="Diagram Batang dari tumor-size")
```

3. Menampilkan diagram lingkaran dari : *tumor-size*

```
tabel.plot.pie(title="Diagram Lingkaran dari tumor-size",autopct='%1.1f%%',
        shadow=True, startangle=90)
```

4. Menampilkan diagram garis dari : *tumor-size*

```
tabel.plot.line(title="Diagram Garis dari tumor-size", linestyle = 'dashed')
```

5. Membuat tabel distribusi

```
data['bins'] = pd.cut(data['deg-malig'], bins=15)
distribusi=data.groupby('bins')['deg-malig'].count()
print(distribusi)
```

histogramnya : 

```
data['deg-malig'].plot.hist(bins=15, facecolor='blue', alpha=0.45)
```

6. Membuat scatter plot

```
data.plot.scatter('age','tumor-size')
```
